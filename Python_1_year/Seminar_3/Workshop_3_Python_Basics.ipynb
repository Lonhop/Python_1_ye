## 0. O(n) vs O(1) — why dictionaries exist

```python
import time

# Searching in a list: O(n)
big_list = list(range(1_000_000))
start = time.perf_counter()
999_999 in big_list
print(f"list search:  {(time.perf_counter()-start)*1000:.2f} ms")

# Searching in a set/dict: O(1)
big_dict = {i: True for i in range(1_000_000)}
start = time.perf_counter()
999_999 in big_dict
print(f"dict search:  {(time.perf_counter()-start)*1000:.3f} ms")
```

A dictionary is a **phone book**: you look up a name (key) and get a number (value) instantly —  
regardless of how many contacts are stored.  
C++ equivalent: `std::unordered_map<K, V>`.

## 1. Dictionary basics

### 1.1 Creating dictionaries

```python
# Literal syntax
student = {"name": "Alice", "grade": 91, "year": 2026}

# From a list of (key, value) pairs
pairs = [("a", 1), ("b", 2), ("c", 3)]
d = dict(pairs)
print(d)

# Empty dict — curly braces (NOT set!)
empty = {}
print(type(empty))   # <class 'dict'>

# Keys can be any hashable type — including tuples!
coords = {(0, 0): "origin", (1, 0): "right", (0, 1): "up"}
print(coords[(0, 0)])
```

### 1.2 Access — `d[key]` vs `d.get(key, default)`

```python
student = {"name": "Alice", "grade": 91}

# Direct access — raises KeyError if missing
print(student["name"])

# Safe access with default — never raises
print(student.get("year", "unknown"))   # 'unknown'
print(student.get("grade", 0))          # 91

# This is equivalent to C++: map.count(key) ? map[key] : default
```

### 1.3 Adding, modifying, deleting

```python
d = {"a": 1, "b": 2}

d["c"] = 3          # add new key
d["a"] = 99         # update existing
print(d)

del d["b"]          # remove key — KeyError if missing
val = d.pop("c")    # remove and return value
print(d, "| popped:", val)

# Update from another dict
d.update({"x": 10, "y": 20})
print(d)
```

### 1.4 Iterating over dictionaries

```python
scores = {"Alice": 91, "Bob": 74, "Carol": 88}

# Keys (default iteration)
for name in scores:
    print(name)

# Values
for score in scores.values():
    print(score)

# Key-value pairs — most useful
for name, score in scores.items():
    print(f"{name}: {score}")
```

### 1.5 Common patterns

```python
words = ["the", "cat", "sat", "on", "the", "mat", "the", "cat"]

# Pattern 1: counter
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1
print(freq)

# Pattern 2: grouping (dict of lists)
lengths = {}
for w in words:
    key = len(w)
    if key not in lengths:
        lengths[key] = []
    lengths[key].append(w)
print(lengths)
```

### 1.6 Dict comprehension

```python
# Basic: {key: value for item in iterable}
squares = {x: x**2 for x in range(1, 6)}
print(squares)   # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# With filter
even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print(even_squares)

# From existing dict — transform values
scores = {"Alice": 91, "Bob": 74, "Carol": 88}
grades = {name: "pass" if score >= 75 else "fail" for name, score in scores.items()}
print(grades)
```

**Task 2.1** — Word frequency  
Count how many times each word appears in the text.  
Print words sorted alphabetically with their counts.

```python
text = "to be or not to be that is the question to be is to do"

d = {}
for word in text.split():
    d[word] = d.get(word, 0) + 1

sorted_dict = sorted(d.items(), key=lambda x: (-x[1], x[0]))
for word, count in sorted_dict:
    print(f"{word}: {count}")
```

```python
d = {"a": 1, "b": 2, "c": 1, "d": 3, "e": 2}

# your code here
grouped = {}
for key, value in d.items():
    if value not in grouped:
        grouped[value] = []
    grouped[value].append(key)
print(grouped)
# expected: {1: ['a', 'c'], 2: ['b', 'e'], 3: ['d']}
```

**Task 2.3** — Most frequent element  
Find the element that appears most often in the list.  
If there's a tie, return the one that appears **first** in the list.  
Use a dict — do not use `Counter` or `.count()`.

```python
lst = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]


# your code here
# expected: 5
def most_frequent(lst):
    freq = {}
    for x in lst:
        freq[x] = freq.get(x, 0) + 1
    most_common = max(freq.items(), key=lambda x: x[1])
    return most_common[0]
print(most_frequent(lst))
```

**Task 2.4** — Group students by grade  
Given a list of `(name, score)` tuples, group students into grade bands:  
`A` ≥ 90, `B` ≥ 75, `C` ≥ 60, `F` below 60.  
Return a dict `{grade: [names]}`, names sorted alphabetically within each group.

```python
results = [
    ("Alice", 91), ("Bob", 74), ("Carol", 88),
    ("Dan", 61), ("Eve", 95), ("Frank", 55), ("Grace", 76)
]
d = {}
for name, score in results:
    if score >= 90:
        d.setdefault("A", []).append(name)
    elif score >= 75:
        d.setdefault("B", []).append(name)
    elif score >= 60:
        d.setdefault("C", []).append(name)
    else:
        d.setdefault("F", []).append(name)
print(d)

# your code here
# expected:
# A: ['Alice', 'Eve']
# B: ['Carol', 'Grace']
# C: ['Bob', 'Dan']
# F: ['Frank']
```

**Task 2.5** — Dict comprehension: normalise scores  
Given a dict of `{name: score}`, build a new dict where each score  
is normalised to the range [0, 1]:  
`normalised = (score - min) / (max - min)`  
Round to 3 decimal places. Use a dict comprehension.

```python
scores = {"Alice": 91, "Bob": 74, "Carol": 88, "Dan": 61, "Eve": 95}

# your code here
# expected: {'Alice': 0.882, 'Bob': 0.382, 'Carol': 0.794, 'Dan': 0.0, 'Eve': 1.0}
max_score = max(scores.values())
min_score = min(scores.values())
normalized = {}
for name, score in scores.items():
    if score != min_score and score != max_score:
        normalized[name] = "{:.3f}".format((score - min_score) / (max_score - min_score))
    else:
        normalized[name] = (score - min_score) / (max_score - min_score)
print(normalized)
```

**Task 3.1** — Index of words  
Build a dictionary that maps each unique word to the **set of positions** it appears at in the text.  
Positions are 0-indexed.

```python
text = "the cat sat on the mat the cat"

# your code here
# expected: {'the': {0, 4, 6}, 'cat': {1, 7}, 'sat': {2}, 'on': {3}, 'mat': {5}}
d = {}
for index in range(len(text.split())):
    word = text.split()[index]
    if word not in d:
        d[word] = set()
    d[word].add(index)
print(d)
```

**Task 3.2** — Two-text diff  
Given two texts, produce a dict with three keys:  
- `"only_a"` — words in text_a but not text_b (sorted list)
- `"only_b"` — words in text_b but not text_a (sorted list)
- `"common"` — words in both (sorted list)

```python
text_a = "machine learning is a subset of artificial intelligence"
text_b = "deep learning is a branch of machine learning and artificial intelligence"

# your code here
lst_a = set(text_a.split())
lst_b = set(text_b.split())
common = (lst_a - lst_b) | (lst_b - lst_a)

d = {
    "only_in_a": lst_a - lst_b,
    "only_in_b": lst_b - lst_a,
    "common": common
}
print(d)
```

**Task 3.3** — Frequency table with rank  
Given a list of words, build a list of `(rank, word, count)` tuples,  
sorted by count descending, then alphabetically for ties.  
Rank starts at 1.

```python
words = ["python", "is", "great", "python", "is", "python", "and", "python", "is", "fast"]

# your code here
# expected:
# (1, 'python', 4)
# (2, 'is', 3)
# (3, 'and', 1)
# (3, 'fast', 1)
# (3, 'great', 1)
d = {}
for word in words:
    d[word] = d.get(word, 0) + 1
sorted_words = sorted(d.items(), key=lambda x: (-x[1], x[0]))
rank = 0
prev_count = None
for i, (word, count) in enumerate(sorted_words, start=1):
    if count != prev_count:
        rank += 1
    print(f"({rank}, '{word}', {count})")
    prev_count = count
```

**Task 3.4** — Sparse matrix  
Represent a sparse matrix (mostly zeros) as a dict with `(row, col)` tuple keys.  
Write functions to:
1. `get(matrix, row, col)` — return value or 0 if absent
2. `set_val(matrix, row, col, val)` — set value (delete key if val is 0)
3. `to_dense(matrix, rows, cols)` — convert to 2D list

```python
def get(matrix: dict, row: int, col: int) -> float:
    # your code here
    return matrix.get((row, col), 0)
    pass

def set_val(matrix: dict, row: int, col: int, val: float):
    # your code here
    matrix[row, col] = val
    pass

def to_dense(matrix: dict, rows: int, cols: int) -> list:
    # your code here
    dense = [[0] * cols for _ in range(rows)]
    for (r, c), val in matrix.items():
        dense[r][c] = val
    return dense


m = {}
set_val(m, 0, 2, 5)
set_val(m, 1, 1, 3)
set_val(m, 2, 0, 7)
print(m)               # {(0,2):5, (1,1):3, (2,0):7}
print(get(m, 1, 1))    # 3
print(get(m, 0, 0))    # 0
for row in to_dense(m, 3, 3):
    print(row)
# [0, 0, 5]
# [0, 3, 0]
# [7, 0, 0]
```

## Summary — All four structures

| Type | Ordered | Mutable | Duplicates | Key type | Typical use |
|---|---|---|---|---|---|
| `list` | yes | yes | yes | index (int) | sequence that changes |
| `tuple` | yes | no | yes | index (int) | fixed record, dict key |
| `set` | no | yes | no | — | unique elements, fast lookup |
| `dict` | yes* | yes | keys: no | any hashable | key→value mapping |

*Dicts preserve insertion order since Python 3.7.*

| Pattern | Code |
|---|---|
| Safe access | `d.get(key, default)` |
| Counter | `freq[w] = freq.get(w, 0) + 1` |
| Grouping | `groups.setdefault(key, []).append(val)` |
| Dict comprehension | `{k: f(v) for k, v in d.items()}` |
| Invert | `{v: k for k, v in d.items()}` |
| Tuple as key | `grid[(row, col)] = value` |
