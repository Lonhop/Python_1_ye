**Task 1.A — LRU Cache (manual)**

Implement a simple LRU (Least Recently Used) cache using a dict.  
The cache has a fixed `capacity`. When full, evict the least recently used key.

```python
cache = LRUCache(2)
cache.put(1, 10)
cache.put(2, 20)
cache.get(1)      # → 10  (1 is now most recently used)
cache.put(3, 30)  # evicts key 2
cache.get(2)      # → -1  (not found)
cache.get(3)      # → 30
```

Hint: Python dicts preserve insertion order — use `dict` + manual reordering.

````python
class LRUCache:
    def __init__(self, capacity: int):
        self.capacity = capacity
        self.cache = {}

    def get(self, key: int) -> int:
        if key not in self.cache:
            return -1
        
        # Move to end (most recently used) by re-inserting
        value = self.cache.pop(key)
        self.cache[key] = value
        return value

    def put(self, key: int, value: int):
        if key in self.cache:
            # Remove existing key to update position
            self.cache.pop(key)
        elif len(self.cache) >= self.capacity:
            # Remove the first item (least recently used)
            # dict maintains insertion order, so first key is LRU
            first_key = next(iter(self.cache))
            self.cache.pop(first_key)
        
        # Insert as most recently used (goes to the end)
        self.cache[key] = value


# Test
cache = LRUCache(2)
cache.put(1, 10)
cache.put(2, 20)
print(cache.get(1))   # → 10  (1 is now most recently used)
cache.put(3, 30)      # evicts key 2
print(cache.get(2))   # → -1  (not found)
print(cache.get(3))   # → 30
````

**Task 1.B — Group anagrams**

Given a list of words, group them into lists of anagrams.  
Return a list of groups (each group sorted), the full list sorted by group size descending.  
No imports — no `Counter`, no `sorted` with key tricks you haven't defined yourself.

```
["eat","tea","tan","ate","nat","bat"]
→ [['ate','eat','tea'], ['nat','tan'], ['bat']]
```

````python
def group_anagrams(words: list) -> list:
    anagram_groups = {}
    
    for word in words:
        chars = list(word)
        for i in range(len(chars)):
            for j in range(0, len(chars) - i - 1):
                if chars[j] > chars[j + 1]:
                    chars[j], chars[j + 1] = chars[j + 1], chars[j]
        sorted_word = ''.join(chars)
        if sorted_word not in anagram_groups:
            anagram_groups[sorted_word] = []
        anagram_groups[sorted_word].append(word)
    for group in anagram_groups.values():
        for i in range(len(group)):
            for j in range(0, len(group) - i - 1):
                if group[j] > group[j + 1]:
                    group[j], group[j + 1] = group[j + 1], group[j]
    result = list(anagram_groups.values())
    for i in range(len(result)):
        for j in range(0, len(result) - i - 1):
            if len(result[j]) < len(result[j + 1]):
                result[j], result[j + 1] = result[j + 1], result[j]
    
    return result

print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
print(group_anagrams([""])) 
print(group_anagrams(["a"]))
````

**Task 1.C — Word ladder distance**

Given two words of the same length and a dictionary of valid words,  
find the **minimum number of single-character substitutions** to get from `start` to `end`,  
where every intermediate word must be in `word_set`.  
Return the number of steps, or -1 if impossible.  
Use BFS — implement it manually with a list as a queue.

```
start="hit", end="cog",
word_set={"hot","dot","dog","lot","log","cog"}
→ 4  (hit→hot→dot→dog→cog)
```

````python
def word_ladder(start: str, end: str, word_set: set) -> int:
    # your code here
    if start == end:
        return 0
    words = set(word_set)
    queue = [(start, 1)]
    visited = {start}
    front = 0
    while front < len(queue):
        current_word, steps = queue[front]
        front += 1
        for i in range(len(current_word)):
            for c in 'abcdefghijklmnopqrstuvwxyz':
                if c == current_word[i]:
                    continue
                next_word = current_word[:i] + c + current_word[i+1:]
                if next_word == end:
                    return steps
                if next_word in words and next_word not in visited:
                    visited.add(next_word)
                    queue.append((next_word, steps + 1))
    return -1

ws = {"hot","dot","dog","lot","log","cog"}
print(word_ladder("hit", "cog", ws))   # 4
print(word_ladder("hit", "xyz", ws))   # -1
````

**Task 1.D — Flatten nested dict**

Given an arbitrarily nested dictionary, flatten it into a single-level dict  
where keys are dot-separated paths.

```python
{"a": {"b": 1, "c": {"d": 2}}, "e": 3}
→ {"a.b": 1, "a.c.d": 2, "e": 3}
```

````python
def flatten(d: dict, prefix: str = "") -> dict:
    # your code here
    result = {}
    for key, value in d.items():
        new_key = f"{prefix}.{key}" if prefix else key
        if isinstance(value, dict):
            result.update(flatten(value, new_key))
        else:
            result[new_key] = value
    return result
    pass


nested = {"a": {"b": 1, "c": {"d": 2}}, "e": 3, "f": {"g": {"h": 4}}}
print(flatten(nested))
# {'a.b': 1, 'a.c.d': 2, 'e': 3, 'f.g.h': 4}
````
