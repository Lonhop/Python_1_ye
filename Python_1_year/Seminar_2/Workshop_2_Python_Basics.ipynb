## 0. Why not always use a list?

````python
# A GPS coordinate — should it ever change?
location = (55.7558, 37.6173)   # Moscow: (lat, lon)

# A list of visited cities — changes over time
visited = ["Moscow", "Berlin", "Paris"]

# The unique countries in a dataset — order doesn't matter, duplicates don't either
countries = {"Russia", "Germany", "France", "Germany", "Russia"}
print(countries)   # no duplicates, order not guaranteed
````

Three different needs → three different types:
- `tuple` — fixed record, immutable, can be used as a dict key
- `list` — ordered sequence that changes
- `set` — unordered collection of unique elements, O(1) membership

## 1. Tuples

### 1.1 Creating tuples

````python
point  = (10, 20)          # typical: fixed-size record
rgb    = (255, 128, 0)
empty  = ()
single = (42,)             # trailing comma is REQUIRED for a single element

# Without the comma it's just parentheses around an int:
not_a_tuple = (42)
print(type(single), type(not_a_tuple))   # tuple vs int
````

### 1.2 Tuple vs List vs C++

| | C++ | Python list | Python tuple |
|---|---|---|---|
| Ordered | yes | yes | yes |
| Mutable | yes | **yes** | **no** |
| Can be dict key | — | no | **yes** |
| Typical use | array / vector | changing sequence | fixed record |
| C++ analogue | `std::vector` | `std::vector` | `std::pair` / `std::tuple` |

Tuples are faster than lists and signal to the reader: *"these elements belong together and won't change."*

### 1.3 Indexing and slicing — same as lists

````python
rgb = (255, 128, 0)
print(rgb[0])      # 255
print(rgb[-1])     # 0
print(rgb[1:])     # (128, 0)
print(len(rgb))    # 3
print(0 in rgb)    # True
````

### 1.4 Unpacking

````python
# Basic unpacking
x, y = (10, 20)
print(x, y)

# Swap — Pythonic one-liner (no temp variable needed)
a, b = 1, 2
a, b = b, a
print(a, b)   # 2 1

# Star unpacking: capture the rest
first, *rest = (1, 2, 3, 4, 5)
print(first, rest)    # 1 [2, 3, 4, 5]

*head, last = (1, 2, 3, 4, 5)
print(head, last)     # [1, 2, 3, 4] 5
````

## 2. Sets

### 2.1 Creating sets

````python
# Curly braces — but NOT empty (that's a dict!)
fruits = {"apple", "banana", "cherry", "apple"}
print(fruits)          # duplicates removed, order not guaranteed

# Empty set — must use set()
empty = set()
print(type(empty))     # <class 'set'>

# From any iterable
letters = set("mississippi")
print(letters)         # unique letters only

nums = set([3, 1, 4, 1, 5, 9, 2, 6, 5])
print(nums)
````

### 2.2 Sets vs C++ unordered_set

| | C++ `unordered_set` | Python `set` |
|---|---|---|
| Lookup | O(1) average | O(1) average |
| Ordered | no | no |
| Duplicates | no | no |
| Mutable | yes | yes |
| Can store | hashable types | hashable types (int, str, tuple) |

> **PEP-8 note:** use `snake_case` for all variable names — `unique_words`, not `UniqueWords`.

### 2.3 Membership and basic operations

````python
primes = {2, 3, 5, 7, 11, 13}

print(7 in primes)    # True  — O(1)
print(4 in primes)    # False
print(len(primes))    # 6

# Add / remove
primes.add(17)
primes.discard(2)     # safe: no error if element absent
# primes.remove(2)    # would raise KeyError if absent
print(primes)
````

### 2.4 Set operations

````python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

print(a | b)   # union        — all elements
print(a & b)   # intersection — common elements
print(a - b)   # difference   — in a but not b
print(a ^ b)   # symmetric diff — in one but not both

# Subset / superset checks
print({1, 2} <= a)   # True: {1,2} is subset of a
print(a >= {1, 2})   # True: a is superset of {1,2}
````

**Task 3.2** — Group by quadrant  
Given a list of `(x, y)` points, count how many fall in each quadrant.  
Ignore points on axes.

````python
points = [(1, 2), (-1, 3), (-2, -1), (4, -3), (2, 5), (-3, -4), (1, -1)]

# your code here
# expected: Q1: 2, Q2: 2, Q3: 2, Q4: 2  (or however many are in your data)
q1 = q2 = q3 = q4 = 0
for x, y in points:
    if x > 0 and y > 0:
        q1 += 1
    elif x < 0 and y > 0:
        q2 += 1
    elif x < 0 and y < 0:
        q3 += 1
    elif x > 0 and y < 0:
        q4 += 1
print(f"Q1: {q1}, Q2: {q2}, Q3: {q3}, Q4: {q4}")
````

**Task 3.3** — Top scorers  
From the list of `(name, score)` tuples, return the **set of names** of students who scored above the average.  
Print average rounded to 1 decimal place and the set of names.

````python
results = [("Alice", 91), ("Bob", 74), ("Carol", 88), ("Dan", 61), ("Eve", 95), ("Frank", 70)]

# your code here
# expected: average: 79.8 | top scorers: {'Alice', 'Carol', 'Eve'}
total_score = sum(score for _, score in results)
average_score = total_score / len(results)
max_score = max(score for _, score in results)
top_scorers = {name for name, score in results if score == max_score}
print(f"average: {average_score:.1f} | top scorers: {top_scorers}")
````

**Task 3.4** — Exclusive words  
Given two texts, find:
- words that appear **only** in `text_a` (not in `text_b`)

Print each as a sorted list.

````python
text_a = "data science is the study of data and information"
text_b = "information theory is the study of communication and signals"

# your code here
# exclusive to text_a: ['data', 'science']  (approximately)
# exclusive to text_b: ['communication', 'signals', 'theory']
set_a = set(text_a.split())
set_b = set(text_b.split())
exclusive_a = set_a - set_b
exclusive_b = set_b - set_a
print(f"exclusive to text_a: {sorted(exclusive_a)}")
print(f"exclusive to text_b: {sorted(exclusive_b)}")
````

## Summary

| Type | Ordered | Mutable | Duplicates | Hashable | Typical use |
|---|---|---|---|---|---|
| `list` | yes | yes | yes | no | sequence that changes |
| `tuple` | yes | **no** | yes | **yes** | fixed record, dict key |
| `set` | no | yes | **no** | no | unique elements, fast lookup |

Key operations at a glance:

| Task | Tool |
|---|---|
| Unpack a record | `a, b, c = my_tuple` |
| Swap variables | `a, b = b, a` |
| Deduplicate | `set(lst)` |
| Intersection | `a & b` |
| Union | `a \| b` |
| Difference | `a - b` |
| Fast membership | `x in my_set` — O(1) |
