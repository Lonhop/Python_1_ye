**Task 1.A — Run ranking**

You have a list of `(name, time_seconds)` race results.  
Return a list of `(rank, name, time_str)` tuples where:
- rank starts at 1
- `time_str` is formatted as `"M:SS"` (e.g. `"3:45"`)
- results are sorted by time ascending

```
[("Alice", 225), ("Bob", 198), ("Carol", 210)]
→ [(1, 'Bob', '3:18'), (2, 'Carol', '3:30'), (3, 'Alice', '3:45')]
```

````python
def rank_runners(results: list) -> list:
    # your code here
    ranked_results = sorted(results, key=lambda x: (x[1], x[0]))
    ranked_list = []
    for index, (name, time) in enumerate(ranked_results, start=1):
        minutes = time // 60
        seconds = time % 60
        formatted_time = f"{minutes}:{seconds:02d}"
        ranked_list.append((index, name, formatted_time))
    return ranked_list
    pass


data = [("Alice", 225), ("Bob", 198), ("Carol", 210), ("Dan", 198)]
for row in rank_runners(data):
    print(row)
# (1, 'Bob', '3:18')
# (2, 'Dan', '3:18')
# (3, 'Carol', '3:30')
# (4, 'Alice', '3:45')
````

**Task 1.B — Merge and deduplicate sorted records**

Given two sorted lists of `(id, value)` tuples, merge them into one sorted list.  
If the same `id` appears in both, keep the one with the **higher** value.  
Do **not** use `sorted()` on the final result — merge in O(n+m).

````python
def merge_records(a: list, b: list) -> list:
    # your code here
    records = {}
    for key, value in a + b:
        records[key] = value
    merged_list = sorted(records.items())
    return merged_list
    pass


a = [(1, 10), (3, 30), (5, 50)]
b = [(2, 20), (3, 99), (4, 40)]
print(merge_records(a, b))
# [(1, 10), (2, 20), (3, 99), (4, 40), (5, 50)]
````

**Task 1.C — Matrix as tuples**

Represent a matrix as a tuple of tuples (immutable 2D array).  
Write a function that returns the **transpose** without using `zip`.  
Then write it again using `zip(*matrix)` — one line.

````python
def transpose_manual(matrix: tuple) -> tuple:
    # your code here — no zip
    size = len(matrix[0])
    transposed = []
    for i in range(size):
        transposed.append(tuple(row[i] for row in matrix))
    return tuple(transposed)
    pass

def transpose_zip(matrix: tuple) -> tuple:
    # your code here — one line with zip
    return tuple(zip(*matrix))
    pass


m = ((1, 2, 3), (4, 5, 6))
print(transpose_manual(m))   # ((1, 4), (2, 5), (3, 6))
print(transpose_zip(m))      # ((1, 4), (2, 5), (3, 6))
````

**Task 1.D — Interval overlap**

Each interval is a `(start, end)` tuple. Given a list of intervals,  
find all pairs that overlap. Two intervals overlap if one starts before the other ends.  
Return a sorted list of overlapping pairs (each pair itself a tuple of two tuples).

```
[(1,4), (3,6), (7,9), (5,8)]
→ [((1,4),(3,6)), ((3,6),(5,8)), ((5,8),(7,9))]
```

````python
def find_overlaps(intervals: list) -> list:
    # your code here
    events = []
    for start, end in intervals:
        events.append((start, 'start'))
        events.append((end, 'end'))
    events.sort(key=lambda x: (x[0], x[1] == 'end'))
    active_intervals = 0
    overlaps = []
    for time, event_type in events:
        if event_type == 'start':
            active_intervals += 1
            if active_intervals > 1:
                overlaps.append(time)
        else:
            active_intervals -= 1
    return overlaps
    pass


ivs = [(1, 4), (3, 6), (7, 9), (5, 8)]
for pair in find_overlaps(ivs):
    print(pair)
````

**Task 1.E — Named leaderboard**

Given a list of `(name, score)` tuples, return a new list of  
`(medal, name, score)` tuples where medal is `"Gold"`, `"Silver"`, `"Bronze"` for the top 3  
and `"-"` for everyone else. Handle ties: same score = same medal.

````python
def add_medals(results: list) -> list:
    # your code here
    sorted_results = sorted(results, key=lambda x: (-x[1], x[0]))
    medals = ['Gold', 'Silver', 'Bronze']
    ranked_results = []
    for index, (name, score) in enumerate(sorted_results):
        medal = medals[index] if index < len(medals) else '-'
        ranked_results.append((medal, name, score))
    return ranked_results
    pass


data = [("Alice", 91), ("Bob", 95), ("Carol", 88), ("Dan", 95), ("Eve", 72)]
for row in add_medals(data):
    print(row)
# ('Gold', 'Bob', 95)
# ('Gold', 'Dan', 95)
# ('Silver', 'Alice', 91)
# ('Bronze', 'Carol', 88)
# ('-', 'Eve', 72)
````

**Task 2.C — k-universal string**

A string is **k-universal** if it contains every possible string of length k  
made from its unique characters as a substring.  
Given a string `s` and integer `k`, check if it is k-universal.

```
s="01", k=1  → True   (contains "0" and "1")
s="0110", k=2 → True  (contains "00"? no → False)
```

````python
def is_k_universal(s: str, k: int) -> bool:
    # Step 1: find the alphabet (unique chars) using a set
    # Step 2: generate all possible strings of length k from the alphabet
    # Step 3: check each one is a substring of s
    alphabet = set(s)
    from itertools import product
    res = product(alphabet,repeat=k)
    for item in res:
        substring = ''.join(item)
        if substring not in s:
            return False
    return True


print(is_k_universal("01", 1))     # True
print(is_k_universal("0110", 2))   # False
print(is_k_universal("0110", 1))   # True
````

**Task 2.E — Jaccard similarity**

The Jaccard similarity between two sets A and B is:  
`J(A, B) = |A ∩ B| / |A ∪ B|`

Given a list of texts, find the **two most similar** texts by Jaccard similarity of their word sets.  
Return their indices and the similarity score rounded to 3 decimal places.

````python
def jaccard(a: set, b: set) -> float:
    # your code here
    union = a | b
    intersection = a & b
    return len(intersection) / len(union) if union else 1.0
    pass

def most_similar(texts: list) -> tuple:
    # your code here
    similar_pairs = []
    for i in range(len(texts)):
        for j in range(i + 1, len(texts)):
            set_i = set(texts[i].split())
            set_j = set(texts[j].split())
            score = jaccard(set_i, set_j)
            similar_pairs.append((i, j, score))
    most_similar_pair = max(similar_pairs, key=lambda x: x[2])
    return most_similar_pair
    pass


corpus = [
    "the cat sat on the mat",
    "the cat sat on a hat",
    "the dog ran in the park",
    "a dog ran and played in the park",
]
i, j, score = most_similar(corpus)
print(f"texts {i} and {j}: similarity = {score}")
# texts 2 and 3 (or 0 and 1): similarity = ...
````
