## 1. Python vs C++: Key Differences

You already know how to program. This section is about *translation*, not learning from scratch.

### 1.1 Syntax at a glance

| Concept | C++ | Python |
|---|---|---|
| Variable | `int x = 5;` | `x = 5` |
| Print | `cout << x << endl;` | `print(x)` |
| Condition | `if (x > 0) { ... }` | `if x > 0:` |
| For loop | `for (int i=0; i<n; i++) { ... }` | `for i in range(n):` |
| Comment | `// comment` | `# comment` |
| Block delimiter | `{ }` | **indentation** |
| Type declaration | required | not needed |
| Line terminator | `;` | newline |

**The most important difference:** Python uses *indentation* to define code blocks. There are no braces. 4 spaces is the standard.

### 1.2 Control flow demo

````python
# if / elif / else
x = 7

if x > 10:
    print("big")
elif x > 5:
    print("medium")
else:
    print("small")
````

````python
# for loop over a range
for i in range(5):
    print(i, end=" ")
````

````python
# for loop directly over a collection — no index needed
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
````

````python
# 'in' operator: membership test
print("banana" in fruits)   # True
print("mango" in fruits)    # False
````

````python
# enumerate: when you need both index and value
for i, fruit in enumerate(fruits):
    print(i, fruit)
````

## 2. Numeric Types

Python has three main numeric types: `int`, `float`, and `complex`. There is also `bool`, which is a subclass of `int`.

### 2.1 Types and type checking

````python
a = 42          # int
b = 3.14        # float
c = True        # bool  (True == 1, False == 0)
d = 2 + 3j      # complex

print(type(a), type(b), type(c), type(d))
````

````python
# Python integers have arbitrary precision — no overflow
print(2 ** 100)
````

### 2.2 Arithmetic operators

| Operator | Meaning | Example | Result |
|---|---|---|---|
| `+` `-` `*` | standard | `7 - 2` | `5` |
| `/` | true division (always float) | `7 / 2` | `3.5` |
| `//` | integer (floor) division | `7 // 2` | `3` |
| `%` | remainder (modulo) | `7 % 2` | `1` |
| `**` | exponentiation | `2 ** 8` | `256` |

````python
print(7 / 2)   # 3.5  — note: NOT 3 like in C++
print(7 // 2)  # 3
print(7 % 2)   # 1
print(2 ** 8)  # 256
````

### 2.3 Type conversion

````python
print(int(3.9))    # 3  — truncates, does NOT round
print(float(5))    # 5.0
print(str(42))     # '42'
print(bool(0))     # False
print(bool(42))    # True
````

### 2.4 Useful built-ins

````python
print(abs(-7))          # 7
print(round(3.14159, 2))  # 3.14
print(pow(2, 10))       # 1024
````

## 3. Strings

Strings in Python are **immutable sequences of characters**. They are one of the most feature-rich built-in types.

### 3.1 Creating strings

````python
s1 = 'hello'
s2 = "world"
s3 = '''multi
line'''
s4 = r"C:\Users\name"   # raw string: backslashes are literal

print(s1, s2)
print(s3)
print(s4)
````

### 3.2 Indexing and slicing

```
 P  y  t  h  o  n
 0  1  2  3  4  5   (positive indices)
-6 -5 -4 -3 -2 -1   (negative indices)
```

Slice syntax: `s[start:stop:step]` — `stop` is **exclusive**

````python
s = "Python"

print(s[0])      # 'P'
print(s[-1])     # 'n'
print(s[1:4])    # 'yth'
print(s[:3])     # 'Pyt'
print(s[::2])    # 'Pto'
print(s[::-1])   # 'nohtyP'  — reverse
````

### 3.3 Key operations and methods

````python
s = "  Hello, World!  "

print(len(s))              # length
print(s.strip())           # remove leading/trailing whitespace
print(s.lower())           # lowercase
print(s.upper())           # uppercase
print(s.replace("World", "Python"))
print(s.find("World"))     # index of first occurrence, -1 if not found
print("llo" in s)          # membership test
````

````python
# split and join — very common in data processing
sentence = "the quick brown fox"
words = sentence.split()         # split on whitespace
print(words)

rejoined = "-".join(words)       # join with separator
print(rejoined)
````

### 3.4 f-strings (formatted string literals)

````python
name = "Alice"
score = 95.678

print(f"Student: {name}")
print(f"Score: {score:.2f}")       # 2 decimal places
print(f"Score: {score:08.2f}")     # zero-padded, width 8
print(f"Double: {score * 2}")
````

## 4. Lists

A **list** is a mutable, ordered sequence of elements. It can hold elements of mixed types.

### 4.1 Creating lists

````python
# Lists: mutable
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
mixed   = [1, "hello", 3.14, True]   # mixed types — unusual in C++
empty   = []

print(type(numbers))   # <class 'list'>
````

### 4.2 Indexing, slicing, membership — same as strings

````python
nums = [10, 20, 30, 40, 50]

print(nums[0])      # 10
print(nums[-1])     # 50
print(nums[1:4])    # [20, 30, 40]
print(30 in nums)   # True
print(len(nums))    # 5
````

### 4.3 Modifying lists

````python
lst = [1, 2, 3]

lst.append(4)          # add to end
lst.insert(1, 99)      # insert at index 1
lst.extend([5, 6])     # add multiple elements
print(lst)

lst.remove(99)         # remove first occurrence of value
popped = lst.pop()     # remove and return last element
print(lst, "| popped:", popped)
````

### 4.4 Sorting

````python
nums = [3, 1, 4, 1, 5, 9, 2, 6]

# sorted() returns a NEW list, original unchanged
print(sorted(nums))
print(nums)             # still unsorted

# .sort() modifies IN PLACE, returns None
nums.sort()
print(nums)             # now sorted

# reverse sort
print(sorted(nums, reverse=True))
````

### 4.5 Useful built-ins for lists

````python
nums = [3, 1, 4, 1, 5, 9, 2, 6]

print(min(nums))    # 1
print(max(nums))    # 9
print(sum(nums))    # 31
print(nums.count(1))  # 2 — how many times 1 appears
````

## 5. List Comprehensions

List comprehensions are a concise, readable way to build lists. They are one of the most Pythonic features of the language.

### 5.1 Basic syntax

```python
[expression for item in iterable]
[expression for item in iterable if condition]
```

Compare with a `for` loop:

````python
# for loop version
squares = []
for x in range(10):
    squares.append(x ** 2)
print(squares)

# list comprehension version
squares = [x ** 2 for x in range(10)]
print(squares)
````

### 5.2 With a filter condition

````python
# Only even squares
even_squares = [x ** 2 for x in range(10) if x % 2 == 0]
print(even_squares)

# Filter words by length
words = ["the", "quick", "brown", "fox", "jumps"]
long_words = [w for w in words if len(w) > 3]
print(long_words)
````

### 5.3 Nested comprehensions

````python
# Flatten a 2D list
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [x for row in matrix for x in row]
print(flat)

# Build a 3x3 identity matrix
n = 3
identity = [[1 if i == j else 0 for j in range(n)] for i in range(n)]
for row in identity:
    print(row)
````

### 5.4 When NOT to use list comprehensions

List comprehensions are great for **transforming or filtering** data in one readable line.  
Avoid them when:
- The logic is complex (multiple conditions, nested branching)
- You have side effects (printing, writing to files)
- The result is hard to read at a glance

**Rule of thumb:** if you can't read it aloud in one breath, use a regular loop.
